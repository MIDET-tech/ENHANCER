###↓記事で用いたコードです。これを実行すればハングマンで遊べます！！

```
def hangman():
    import random as ran
  #ステージの定義
    stages = ["",
            "__________           ",
            "|         |          ",
            "|         0          ",
            "|        /|\         ",
            "|        / \         ",
            "|                    "
            ]
    wrong = 0
  #正解の単語のリストの定義
    lis = ["cat", "dog", "penguin", "bird", "fish", "panda"]
  #正解のリストからランダムに単語を選ぶ
    num = ran.randint(0, 5)
    word = lis[num]
  #correctという変数に、正解の単語をリストとして代入
    correct = list(word)
  #boardという変数に、正解の単語の数だけ＿を代入
    board = ["_"] * len(word)
    win = False
    print("ハングマンへようこそ！")
  #ハングマンのゲームを進めていくループ処理
    while wrong < len(stages) - 1:
        print("\n")
        msg = "1文字を予想してね! ヒント:動物の名前"
        char = input(msg)
      #回答の文字が正解の単語に含まれていた場合に動作
        if char in correct:
            cind = correct.index(char)
          #回答の文字を表示
            board[cind] = char
          #正解単語の中の解答の文字の部分を$に置き換える
            correct[cind] = '$'
      #回答の文字が間違っていた場合に動作
        else:
            wrong += 1
      #正解の単語の文字数分の＿を表示
        print(" ".join(board))
      #間違った数だけハングマンの絵を表示
        e = wrong + 1
        print("\n".join(stages[0 : e]))
      #勝った場合の動作
        if "_" not in board:
            print("あなたの勝ち！")
            print(" ".join(board))
            win = True
            break
  #負けた場合の動作
    if not win:
        print("\n".join(stages[0 : wrong + 1]))
        print("あなたの負け！正解は{}". format(word))
            
hangman()
```
