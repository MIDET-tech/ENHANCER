```
x = open("x.txt", "w")
x.write("hello, world")
x.close()
```

```
with open("x.txt", "w")as x:
  x.write("hello, world")
```

```
with open("x.txt", "r") as x:
  print(x.read())
```

```
with open("y.txt", "w", encoding="utf-8") as y:
  y.write("こんにちは")
with open("y.txt", "r", encoding="utf-8") as y:
  print(y.read())
```
