#####ユーザー定義関数②

```
def f(x): 
  return x * 3 

e = f(2)

if e == 6:
  print("e is 6")
else:
  print("e is not 6")
```

####スコープ

####グローバルスコープ（グローバル変数）

```
#グローバルスコープ（グローバル変数）
x = 1
y = 2
z = 3
def my():
  print(x)
  print(y)
  print(z)

my()
```

#####ローカルスコープ（ローカル変数）

```
def your():
  p = 1
  e = 2
  n = 3

print(p)
print(e)
print(n)
```

```
#ローカルスコープ（ローカル変数）
def your():
  p = 1
  e = 2
  n = 3
  print(p)
  print(e)
  print(n)

your()
```

####グローバル変数を関数内から呼ぶ

```
#qをグローバルスコープ（変数）として定義
q = 2

#関数を定義
def  gf():
  global q
  q -= 1
  print(q)

gf()
```
