####mathモジュール

```
#mathモジュールをインポート
import math

#mathの中のpow関数を用いる
math.pow(2, 3)
```

####randomモジュール

```
#randomモジュールをインポート
import random

#指定した範囲の乱数を生成
random.randint(0, 500)
```

####statisticsモジュール

```
#statisticsモジュールをインポート
import statistics

number = [1, 50, 4, 22, 4, 18, 27]
#平均値（mean）
statistics.mean(number)
```

```
#中央値（median）
statistics.median(number)
```

```
#最頻値（mode）
statistics.mode(number)
```

####別の文字列としてモジュールをインポートする

```
#statisticsをstとしてインポート
import statistics as st

#平均値（mean）
st.mean(number)
```
